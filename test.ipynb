```python
def corrigir_data(data):
    partes_data = data.split('/')
    dia, mes, ano = map(int, partes_data)
    
    # Verifica se o mês está correto
    if mes > 12:
        mes = int(str(mes)[0])
    
    # Verifica se o dia cabe no mês
    if dia > 31:
        dia = int(str(dia)[0])
    
    data_corrigida = f"{dia}/{mes}/{ano}"
    return data_corrigida

# Exemplo de uso
data_errada = "22/20/2023"
data_corrigida = corrigir_data(data_errada)
print(data_corrigida)  # Saída: 2/2/2023
```

```python
import pandas as pd
# Exemplo de DataFrame com uma coluna de datas
data = {'Data': ['2024-01-15', '2024-06-20', '2024-12-31']}
df = pd.DataFrame(data)

# Converter a coluna de datas para o tipo datetime
df['Data'] = pd.to_datetime(df['Data'])

# Extrair o nome do mês abreviado em português (Jan, Fev, Mar, etc.)
df['NomeMes'] = df['Data'].dt.strftime('%b')

# Extrair o ano no formato YYYY (ex: 2024)
df['Ano'] = df['Data'].dt.strftime('%Y')

print(df)
```
